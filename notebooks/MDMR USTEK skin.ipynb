```r
TO_DIST<-function(x){
  x<-x[,2:ncol(x)]
  rownames(x)<-colnames(x)
  x<-as.matrix(x)
  as.dist(x)
}

BRAY.DIST<-TO_DIST(BRAY)
JACC.DIST<-TO_DIST(JACC)
U.UNIFRAC.DIST<-TO_DIST(U.UNIFRAC)
W.UNIFRAC.DIST<-TO_DIST(W.UNIFRAC)

BRAY40.DIST<-TO_DIST(BRAY40)
JACC40.DIST<-TO_DIST(JACC40)
U.UNIFRAC40.DIST<-TO_DIST(U.UNIFRAC40)
W.UNIFRAC40.DIST<-TO_DIST(W.UNIFRAC40)
```

```r
# adjust p values for model where week is continous variable
PVALS<-as.numeric()

PVALS[1]<-mdmr.res.brayt[c(3),3]
PVALS[2]<-mdmr.res.jacct[c(3),3]
PVALS[3]<-mdmr.res.uunifract[c(3),3]
PVALS[4]<-mdmr.res.wunifract[c(3),3]

#QVAL<-qvalue(PVALS,lambda = seq(0, max(PVALS), 0.05))$qvalues
FDR<-p.adjust(PVALS,"fdr")
FDR
```

```r
# adjust p values for model where week is categorical variable
PVALSC<-as.numeric()

PVALSC[1]<-mdmr.res.braytc[c(3),3]
PVALSC[2]<-mdmr.res.jacctc[c(3),3]
PVALSC[3]<-mdmr.res.uunifractc[c(3),3]
PVALSC[4]<-mdmr.res.wunifractc[c(3),3]

#QVAL<-qvalue(PVALS,lambda = seq(0, max(PVALS), 0.05))$qvalues
FDRC<-p.adjust(PVALSC,"fdr")
FDRC
```
