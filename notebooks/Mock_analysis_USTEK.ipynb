```python
import pandas as pd
```

### Mock community standard

```python
mcs = pd.read_excel("/Users/dagmarschierova/MEGA/MBÚ/USTEK/output/mock/taxonomy/Microbial_Community_Standard_I.xlsx", engine="openpyxl")
mcs.head()
```

```python
mock = mcs[["mock", "relative abundance"]].set_index("mock").dropna()
newnbgg99 = mcs[["new nb gg 99", "new nb gg 99 rel abund"]].set_index("new nb gg 99").dropna()
newnbsilva99 = mcs[["new nb silva 99", "new nb silva 99 rel abund"]].set_index("new nb silva 99").dropna()
pretrainednbgg99 = mcs[["pretrained nb gg 99", "pretrained nb gg 99 rel abund"]].set_index("pretrained nb gg 99").dropna()
pretrainednbsilva99 = mcs[["pretrained nb silva 99", "pretrained nb silva 99 rel abund"]].set_index("pretrained nb silva 99").dropna()
vsearchgg99 = mcs[["vsearch gg 99", "vsearch gg 99 rel abund"]].set_index("vsearch gg 99").dropna()
vsearchsilva99 = mcs[["vsearch silva 99", "vsearch silva 99 rel abund"]].set_index("vsearch silva 99").dropna()
```

```python
mcs_joined = mock.join(vsearchsilva99, how='outer')
print(len(mcs_joined.index))
mcs_joined = mcs_joined.join(vsearchgg99, how='outer')
print(len(mcs_joined.index))
mcs_joined = mcs_joined.join(pretrainednbsilva99, how='outer')
print(len(mcs_joined.index))
mcs_joined = mcs_joined.join(pretrainednbgg99, how='outer')
print(len(mcs_joined.index))
mcs_joined = mcs_joined.join(newnbgg99, how='outer')
print(len(mcs_joined.index))
mcs_joined = mcs_joined.join(newnbsilva99, how='outer')
print(len(mcs_joined.index))
```

```python
mcs_joined = mcs_joined.sort_values(by=['relative abundance'])
mcs_joined
```

```python
mcs_joined.to_csv('/Users/dagmarschierova/MEGA/MBÚ/USTEK/output/mock/taxonomy/Microbial_Community_Standard_I_joined.tsv', sep = '\t', index=True)
```
