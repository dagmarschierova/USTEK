```python
metadata = pd.read_csv("C:\\Users\\Dasa\\Documents\\MEGA\\MBÚ\\swabs_ZS\\inputs\\metadata_swabs.tsv", sep='\t')
metadata = metadata.iloc[1:]
metadata.rename(columns={"#SampleID": "SampleID"}, inplace=True)
metadata.set_index('SampleID', inplace=True)

shannon = pd.read_csv("C:\\Users\\Dasa\\Documents\\MEGA\\MBÚ\\swabs_ZS\\outputs\\qiime2_results_290230\\diversity-D_UST\\alpha\\shannon_tabulated.tsv", sep='\t')
shannon = shannon.iloc[1:]
shannon.set_index('SampleID', inplace=True)

metadata = metadata.join(shannon, how='inner', on='SampleID')
metadata
```

```python
md = smf.mixedlm("shannon_entropy ~ Week_num", metadata, groups=metadata["Patient"], re_formula="~Week_num")
mdf = md.fit(method=["lbfgs"])
print(mdf.summary())
```
